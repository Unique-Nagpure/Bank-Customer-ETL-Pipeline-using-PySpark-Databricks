### Data Quality Report

```python
df_clean = spark.table("testdatabricksuni.uniqueschema.bank_customer_stage4")
```

```python
from pyspark.sql.functions import sum, count, col

report = df_clean.select(

    count("*").alias("Total Rows"),

    sum(col("email_valid").cast("int")).alias("Valid Emails"),

    sum((~col("email_valid")).cast("int")).alias("Invalid Emails"),

    sum(col("phone_valid").cast("int")).alias("Valid Phones"),

    sum((~col("phone_valid")).cast("int")).alias("Invalid Phones")

)

display(report)
```

```python
df_clean.write.mode("overwrite").saveAsTable(
    "testdatabricksuni.uniqueschema.bank_customer_stage5"
)
```
