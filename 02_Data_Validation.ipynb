### Data Validation

```python
from pyspark.sql.functions import col

df = spark.table("testdatabricksuni.uniqueschema.bank_customer_stage1")
```

```python
print(f"Total Rows : {df.count():,}")
print(f"Total Columns : {len(df.columns)}")
```

```python
df.printSchema()
```

```python
duplicate_customers = (
    df
    .groupBy("customer_id")
    .count()
    .filter(col("count") > 1)
)

print("Duplicate Customers :", duplicate_customers.count())

display(duplicate_customers)
```

```python
from pyspark.sql.functions import when, count

null_df = df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
])

display(null_df)
```

```python
display(df.select("gender").distinct())

display(df.select("state").distinct())

display(df.select("account_type").distinct())

display(df.select("city").distinct())

display(df.select("branch").distinct())
```
